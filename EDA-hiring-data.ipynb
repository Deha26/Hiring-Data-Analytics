```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.boxplot(df['Offered Salary'])
plt.title("Outliers in Offered Salary")
plt.ylabel("Salary")
plt.show()
```

```python
Q1 = df['Offered Salary'].quantile(0.25)
Q3 = df['Offered Salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['Offered Salary'] < lower) | (df['Offered Salary'] > upper)]
```

```python
outliers
```

```python
sns.barplot(x='Post Name',y='Offered Salary',data=data)
```

```python
sns.barplot(x='Department',y='Offered Salary',data=data)
plt.xticks(rotation=90)  
plt.tight_layout()      
plt.show()
```

Since the variability of General Management dept is high, the outliers could mean the role for senior executives.
